# Objective: 

The objective of this internship assignment is to create a solution that can automatically generate multiple-choice objective questions based on a given chapter from a subject. The questions should have more than one correct answer to increase complexity and challenge. The aim is to develop a robust solution that aids educators in creating engaging and challenging assessments for students.

```python
import PyPDF2
import random
import os
```

# Created a function

```python
# To generate MCQ

def get_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
        return text

# now creating function to get_mca_questions

def get_mca_questions(context):
    sentences = context.strip().split('. ') # Here splitting the context into sentences
    
    mca_questions = [] # Creating a empty list to store the required MCQ's
    for sentence in sentences:
        question = 'Which of the following is true about ' + sentence + '?' # we can give question type as per the required condition
        choices = random.sample(sentences, k=3) # it'll select random choices
        choices.append(sentence) # it'll add the correct answer to the choices
        random.shuffle(choices)  # it'll shuffle the choices
        
        mca_question = question + '\n'
        for i, choice in enumerate(choices):
            mca_question += chr(ord('a') + i) + ') ' + choice + '\n'
            
        mca_questions.append(mca_question)
        
    return mca_questions

# Here we are going to load our required pdf file to generate the MCQ's. We can add different pdf file from which we can generate the MCQ's.
# First we going to load the Chapter - 2pdf file to generate the MCQ's.

file_path = '/content/chapter-2.pdf'
context = get_text_from_pdf(file_path)

questions = get_mca_questions(context)
for i, question in enumerate(questions):
    print(f'Question {i+1}: {question}')
```
